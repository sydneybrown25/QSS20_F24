- Year: 2015
- grouping variable: GENDER

Print the output in dataframe format and briefly interpret; what do scores look like between the genders?

```python
# Create query
math_2015_query = (
    'https://www.nationsreportcard.gov/'
    'Dataservice/GetAdhocData.aspx?'
    'type=data&subject=mathematics&grade=8&'
    'subscale=MRPCM&variable=GENDER&'
    'jurisdiction=CA&stattype=MN:MN&'
    'Year=2015')

# call API
math_2015_resp = requests.get(math_2015_query)

# get json of status and result
math_2015_resp_j = math_2015_resp.json()

# extract contents in result key and convert to dataframe
math_2015_resp_d = pd.DataFrame(math_2015_resp_j['result'])

# view results
math_2015_resp_d
```

## 1.4 Reproduce the queries from 1.1 and 1.2 using a user-defined function (4 points)

Create a function, `construct_naep_query` that takes in two arguments:

- year: this should be a list with all years (so if one year, single element list; if multiple years, list with those years)
- place: this should be a string with the name of the state or jurisdiction to pull 
    
Have the function return the query and make sure it's identical to the queries you wrote for 1.1 and 1.2 (can use assert or other checker)

```python
def construct_naep_query(year, place):
    if not isinstance(year, list):
        raise ValueError("Year must be provided as a list of years.")
    
    year_str = ','.join(map(str, year))
    
    query = (
        'https://www.nationsreportcard.gov/'
        'Dataservice/GetAdhocData.aspx?'
        'type=data&subject=mathematics&grade=8&'
        'subscale=MRPCM&variable=GENDER&'
        f'jurisdiction={place}&stattype=MN:MN&'
        f'Year={year_str}'
    )
    
    return query    
```

```python
print(construct_naep_query(['2015'], 'CA'))
print(construct_naep_query(['2013','2017'], 'CA'))
```

```python
# Should print nothing if the two queries match 

# 1.1 check
assert construct_naep_query(['2015'], 'CA') == math_2015_query, "They do not match"

# 1.2 check
assert construct_naep_query(['2013', '2015', '2017', '2019'], 'CA') == math_query, "They do not match"
```
